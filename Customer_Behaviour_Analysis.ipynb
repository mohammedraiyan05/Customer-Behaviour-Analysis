```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

# Load the Dataset

```python
import pandas as pd
df = pd.read_csv("/content/retail_data.csv")
df
```

# View Basic Information

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['gender'] = df['gender'].str.title()
df['income_bracket'] = df['income_bracket'].str.title()
```

# Feature Engineering

Create Age Groups

```python
df['age_group'] = pd.cut(df['age'],
                         bins=[18,30,45,60,100],
                         labels=['Young','Adult','Middle-Aged','Senior'])
df.head()
```

Create Family Size Feature

```python
df['family_size'] = df['number_of_children'] + (df['marital_status'] == "Married").astype(int)
df
```

Encode Yes/No to Binary

```python
df['loyalty_program_flag'] = df['loyalty_program'].map({'Yes':1,'No':0})
df
```

# Exploratory Data Analysis (EDA)

Gender Distribution

```python
def add_values(ax):
    for p in ax.patches:
        value = p.get_height()
        ax.text(p.get_x() + p.get_width()/2,
                value + 0.01,
                f"{value:.2f}",
                ha='center', fontsize=10)
```

```python
ax = sns.countplot(x='gender', data=df)
add_values(ax)
plt.title("Gender Distribution")
plt.show()
```

Churn Rate by Gender

```python
ax = sns.barplot(x='gender', y='churned', data=df)
add_values(ax)
plt.title("Churn Rate by Gender")
plt.show()
```

Churn by Age Group

```python
ax = sns.barplot(x='age_group', y='churned', data=df)
add_values(ax)
plt.title("Churn by Age Group")
plt.show()
```

Churn by Income Group

```python
ax = sns.barplot(x='income_bracket', y='churned', data=df)
add_values(ax)
plt.title("Churn by Income Group")
plt.show()
```

Loyalty Program vs Churn

```python
ax = sns.barplot(x='loyalty_program', y='churned', data=df)
add_values(ax)
plt.title("Loyalty Program Effect on Churn")
plt.show()
```

Membership Years vs Churn

```python
sns.boxplot(x='churned', y='membership_years', data=df)
plt.title("Membership Years & Churn")
plt.show()
```

Education Level Distribution

```python
ax = sns.countplot(x='education_level', data=df)
plt.xticks(rotation=45)
add_values(ax)
plt.title("Education Level Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Helper function for value labels
def add_values(ax):
    for p in ax.patches:
        value = p.get_height()
        ax.text(p.get_x() + p.get_width()/2,
                value + 0.02,
                f"{value:.2f}" if value < 1 else int(value),
                ha='center', fontsize=9)

# Create figure
plt.figure(figsize=(18, 12))

# -------------------------
# 1. Gender Distribution
# -------------------------
plt.subplot(2, 3, 1)
ax1 = sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
add_values(ax1)

# -------------------------
# 2. Churn Rate by Gender
# -------------------------
plt.subplot(2, 3, 2)
ax2 = sns.barplot(x='gender', y='churned', data=df)
plt.title("Churn by Gender")
add_values(ax2)

# -------------------------
# 3. Churn by Age Group
# -------------------------
plt.subplot(2, 3, 3)
ax3 = sns.barplot(x='age_group', y='churned', data=df)
plt.title("Churn by Age Group")
add_values(ax3)

# -------------------------
# 4. Churn by Income Group
# -------------------------
plt.subplot(2, 3, 4)
ax4 = sns.barplot(x='income_bracket', y='churned', data=df)
plt.title("Churn by Income Group")
add_values(ax4)

# -------------------------
# 5. Loyalty Program vs Churn
# -------------------------
plt.subplot(2, 3, 5)
ax5 = sns.barplot(x='loyalty_program', y='churned', data=df)
plt.title("Loyalty Program vs Churn")
add_values(ax5)

# -------------------------
# 6. Education Level Distribution
# -------------------------
plt.subplot(2, 3, 6)
ax6 = sns.countplot(x='education_level', data=df)
plt.title("Education Level Distribution")
plt.xticks(rotation=45)
add_values(ax6)

plt.tight_layout()
plt.show()
```
